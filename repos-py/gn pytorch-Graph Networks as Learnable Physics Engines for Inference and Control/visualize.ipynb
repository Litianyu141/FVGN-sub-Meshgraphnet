```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.load('./swimmer2.npy')
```

```python
def decode(data):
    pos = data[10:28].reshape(6,3)
    vel = data[28:].reshape(6,3)
    return pos, vel
```

```python
pos = None
vel = None
for frame in range(0,40,10):
    pos, vel = decode(data[1,frame])
    #print(pos, vel)
    plt.figure()
    for i in range(6):
        angle = pos[i,2]
        x = pos[i,0]
        y = pos[i,1]
        r = 0.05
        dy = np.cos(angle) * r
        dx = - np.sin(angle) * r
        #plt.figure()
        plt.plot([x - dx,x + dx], [y - dy,y + dy])
        plt.axis('equal')
```
